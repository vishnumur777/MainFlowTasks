## **_1. Import packages_**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## **_2. Import dataset_**

```python
df = pd.read_csv("householdtask3.csv")
```

```python
df
```

## **_3. Scatter Plot_**

```python
plt.scatter(df['year'],df['income'])
plt.title('Scatter Plot')
plt.show()
```

## **_4. Scatter Plot ( colorbox )_**

```python
plt.scatter(df['year'],df['expenditure'],c=df['income'])
plt.title('line chart')
plt.colorbar()
plt.show()
```

## **_5. Line Chart_**

```python
plt.plot(df['income'])
plt.plot(df['expenditure'])
plt.xlabel("Year")
plt.ylabel("Expenditure")
plt.title("Line Plot")
plt.show()
```

## **_6. Bar Chart_**

```python
plt.bar(df.year,df.income)
plt.title("Bar Chart")
plt.xlabel("Year")
plt.ylabel("Income")
```

## **_7. Histogram_**

```python
plt.hist(df.income,bins=10)
```
