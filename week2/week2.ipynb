# **_Task 2 - Describe the dataset and clean the Null values_**

### **_1. Importing Libraries_**

```python
import pandas as pd
import numpy as np
```

### **_2. Import dataset_**

```python
df = pd.read_csv("data.csv")
```

```python
df
```

```python
df.head(5)
```

### **_3. Checking Null Values_**

```python
df.isna().sum()
```

### **_4. Column name space removal_**

```python
df.columns = [x.strip() for x in df.columns]
```

### **_5. Filling NaN values for column mean_**

```python
for i in df.columns[1:]:
    df[i].fillna(df[i].mean().round(3),inplace=True)
```

```python
df.isna().sum()
```

### **_6. Displaying information of columns_**

```python
df.info()
```

### **_7. Describing data with the statistical values_**

```python
df.describe()
```

### **_8. Dropping duplicate rows in the dataset_**

```python
df1=df.drop_duplicates()
```

```python
df1
```

### 9. **_Drop observation row which is not relevant datatype (datetime)_**

```python
df2=df1.drop(['Observation'],axis=1)
```

```python
df2
```

### **_10. Calculating Inter Quartile Range_**

```python
Q1 = df.quantile(0.25,numeric_only=True)
Q3 = df.quantile(0.75,numeric_only=True)
IQR = Q3 - Q1
```

```python
IQR
```

### **_11. Boxplot for all the columns_**

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
for i in df2.columns:
    sns.boxplot(df1[i])
    plt.title(i)
    plt.show()
```

### **_12. Removing Outliers_**

```python
threshold = 1.5 * IQR
outliers = (df2 < (Q1 - threshold)) | (df2 > (Q3 + threshold))
```

```python
df2=df2[~outliers.any(axis=1)]
```

```python
df2
```

```python
df2.shape
```

After removing outliers, our dataset has 116 rows and 22 columns.
